```python
import pandas as pd
file_path = '/content/drive/MyDrive/Intro to cloud/questions.json'

with open(file_path, 'r') as file:
    df = pd.read_json(file_path)

# Display the DataFrame
df
```

```python
import json
import random
def get_random_question():
    return random.choice(df["questions"])


random_question = get_random_question()

html_form = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <title>Random Question Form</title>
        <style>
            body{{
                background-image: url("https://i.imgur.com/rOVqQv7.jpeg");
                background-size: cover;
                background-position: center;
                background-repeat: no-repeat;
            }}
            .question{{
              color: black;
            }}
            .question h2{{
              color: black;
            }}
        </style>
        <script>
            function nextQuestion() {{
              var q1 = document.getElementById("q1");
              var q2 = document.getElementById("q2");
              var q3 = document.getElementById("q3");
              var q4 = document.getElementById("q4");
              var q5 = document.getElementById("q5");
              q1.style.display = "none";
              q2.style.display = "none";
              q3.style.display = "none";
              q4.style.display = "none";
              q5.style.display = "none";


              var randomNumber = getRandomNumber(1, 5);
              if(randomNumber == 1)
                q1.style.display = "block";
              if(randomNumber == 2)
                q2.style.display = "block";
              if(randomNumber == 3)
                q3.style.display = "block";
              if(randomNumber == 4)
                q4.style.display = "block";
              if(randomNumber == 5)
                q5.style.display = "block";
            }}
            function getRandomNumber(min, max) {{
                return Math.floor(Math.random() * (max - min + 1)) + min;
            }}

            function validateForm(index) {{
                var selectedOption = document.querySelector('input[name="answer"]:checked');

                //if no option is selecteed
                if (selectedOption == false) {{

                    alert("Please select an answer.");
                    return false;
                }}

                var userAnswer = selectedOption.value;
                var correctAnswer = 0
                if(index == 1)
                  correctAnswer = "{df['questions'][0]['correct_option']}";
                if(index == 2)
                  correctAnswer = "{df['questions'][1]['correct_option']}";
                if(index == 3)
                  correctAnswer = "{df['questions'][2]['correct_option']}";
                if(index == 4)
                  correctAnswer = "{df['questions'][3]['correct_option']}";
                if(index == 5)
                  correctAnswer = "{df['questions'][4]['correct_option']}";


                if (userAnswer == correctAnswer) {{
                    alert("Correct answer!");
                    nextQuestion();  // refresh with new question
                }} else {{
                    alert("Incorrect answer.");
                }}
                return false; //dont submit form
            }}
            function resetForm() {{
                document.getElementById("question-form").reset();
            }}
        </script>
    </head>
    <body>
        <div class="question" id="q1">
            <h2>{df["questions"][0]['question']}</h2>
            <form onsubmit="return validateForm(1);">
                {"".join(f"<input type='radio' name='answer' value='{option}'> {option}<br>" for option in df["questions"][0]['options'])}
                <br>
                <input type="submit" value="Submit">
            </form>
        </div>
        <div class="question" id="q2">
            <h2>{df["questions"][1]['question']}</h2>
            <form onsubmit="return validateForm(2);">
                {"".join(f"<input type='radio' name='answer' value='{option}'> {option}<br>" for option in df["questions"][1]['options'])}
                <br>
                <input type="submit" value="Submit">
            </form>
        </div>
        <div class="question" id="q3">
            <h2>{df["questions"][2]['question']}</h2>
            <form onsubmit="return validateForm(3);">
                {"".join(f"<input type='radio' name='answer' value='{option}'> {option}<br>" for option in df["questions"][2]['options'])}
                <br>
                <input type="submit" value="Submit">
            </form>
        </div>
        <div class="question" id="q4">
            <h2>{df["questions"][3]['question']}</h2>
            <form onsubmit="return validateForm(4);">
                {"".join(f"<input type='radio' name='answer' value='{option}'> {option}<br>" for option in df["questions"][3]['options'])}
                <br>
                <input type="submit" value="Submit">
            </form>
        </div>
        <div class="question" id="q5">
            <h2>{df["questions"][4]['question']}</h2>
            <form onsubmit="return validateForm(5);">
                {"".join(f"<input type='radio' name='answer' value='{option}'> {option}<br>" for option in df["questions"][4]['options'])}
                <br>
                <input type="submit" value="Submit">
            </form>
        </div>
        <script>
            nextQuestion()
        </script>
    </body>
    </html>
    """

from IPython.core.display import display, HTML
display(HTML(html_form))

```
