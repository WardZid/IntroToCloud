```python

#@title The following fucntions, for getting the pages from the DB, sorting them by variable "count" and saving each action for a the user in specific session
import concurrent.futures

def fetch_page_data(hashed_url):
    page_data = fbConn.get('/pages/' + hashed_url, None)
    return (hashed_url, page_data) if page_data else None

def sort_pages(pages_to_sort):
    sorted_pages = []
    for key, page_info in pages_to_sort.items():
        path = 'pointed/' + key + '/pointed_from'
        array_data = fbConn.get(path, None)
        length = len(array_data) if array_data else 0
        count = length if length > 0 else 1
        page_info['count'] *= count
        sorted_pages.append((key, page_info['count']))

    # Sort pages by count in descending order
    sorted_pages = sorted(sorted_pages, key=lambda x: x[1], reverse=True)
    return [page[0] for page in sorted_pages]

def get_indexed_pages(query, session):
    indexed_pages = fbConn.get('/index/' + query, None)
    link_hashes = []
    if indexed_pages:
        link_hashes = sort_pages(indexed_pages)

    session_data = {
        "action": "search",
        "datetime": datetime.now(),
        "input": query,
        "output": link_hashes
    }
    fbConn.post("/sessions/" + session, data=session_data)

    # Fetch indexed pages concurrently
    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_hash = {executor.submit(fetch_page_data, hashed_url): hashed_url for hashed_url in link_hashes}
        for future in concurrent.futures.as_completed(future_to_hash):
            hashed_url = future_to_hash[future]
            try:
                data = future.result()
                if data:
                    link_hashes.append(data)
            except Exception as e:
                print(f"Error fetching indexed page for hashed URL {hashed_url}: {e}")

    return link_hashes

def get_pages_info(hashed_urls):
  links = []

  # Iterate over each hashed URL
  for hashed_url in hashed_urls:
    if hashed_url is not None:
    # Retrieve the URL from the pages node using the hashed URL
      page_data = fbConn.get('/pages/' + hashed_url, None)
      if page_data:
        links.append(page_data)

  if links:
    return links
  else:
    return []
```

```python
# @title function for handling the Sign up operation
def hash_user(user,password):
  posthash = user+password
  return hashlib.md5(posthash.encode()).hexdigest()

def sign_up(username,password):
  hashed_user = hash_user(username,password)
  data_to_upload = {
    "username": username,
    "password": password
  }
  result = fbConn.put('/users',hashed_user,data=data_to_upload)
  print(result)


```

```python
# @title function for handling the Sign in operation
def hash_user(user,password):
  prehash = user+password
  return hashlib.md5(prehash.encode()).hexdigest()

def sign_in(username,password):
  hashed_user = hash_user(username,password)

  user_data = fbConn.get('/users/' + hashed_user, None)
  if user_data is not None:
    session_data={
        "user": hashed_user,
        "start_time": datetime.now()
    }
    result = fbConn.post("sessions", data=session_data)
    return result['name']
  else:
    return None
```

```python
#@title the following fucntions for getting the statistics from the DB

def get_index_data():
    data = fbConn.get('index',None)
    #keys = list(data.keys()) if data else []
    return data

def getAverageFrequencyData():
  word_data = get_index_data()
  average_frequency = {}
  for word, links in word_data.items():
    total_count = sum(count_dict['count'] for count_dict in links.values())
    num_links = len(links)
    average_frequency[word] = round(total_count / num_links)

  chart_data = []
  for word, frequency in average_frequency.items():
    chart_data.append({'x': len(word_data[word]), 'y':frequency , 'label': word})

  return chart_data

def getTotalFrequencyData():
  word_data = get_index_data()
  total_frequency = {}
  for word, links in word_data.items():
    total_frequency[word] = sum(count_dict['count'] for count_dict in links.values())

  chart_data = []
  for word, frequency in total_frequency.items():
    chart_data.append({'x': len(word_data[word]) , 'y':frequency , 'label': word})

  return chart_data
```

```python
#@title <h1>Improved version of indexing and reteriving links functions ( with cache mechanism)</h1><ol><li><strong>sort_pages(pages_to_sort):</strong> This function sorts the pages based on their count values retrieved from Firestore. It iterates through each page in the given dictionary, calculates the count by checking the number of pointed links, and sorts the pages by count in descending order.</li><li><strong>get_indexed_pages1(query, session):</strong> This function retrieves indexed pages from Firestore based on the given query. It first checks if the query result is cached and returns the cached data if available and not expired. If not cached or expired, it fetches the data from Firestore, sorts the pages, updates the cache, and adds the query and response to a session.</li><li><strong>get_pages_info1(hashed_urls):</strong> This function retrieves pages information from Firestore based on the given hashed URLs. It iterates through each hashed URL, retrieves the page data from Firestore, and appends it to the links list. The function then returns the list of page information.</li></ol>


import requests
from collections import deque
from datetime import datetime

# Define constants for Firestore connection and other configurations
FIRESTORE_URL = 'https://braudecloud-18-02-2024-default-rtdb.europe-west1.firebasedatabase.app/'
CACHE_EXPIRATION = 2592000  # Cache expiration time in seconds (30 days)

# Define a cache dictionary to store query results
cache = {}

# Function to fetch pages from Firestore and sort them
def sort_pages(pages_to_sort):
    sorted_pages = []
    for key, page_info in pages_to_sort.items():
        count = 0
        array_data = fbConn.get('pointed/' + key + '/pointed_from', None)
        if array_data:
            count = len(array_data)
        page_info['count'] *= count
        sorted_pages.append((key, page_info['count']))

    # Sort pages by count in descending order
    sorted_pages = sorted(sorted_pages, key=lambda x: x[1], reverse=True)
    #return [page[0] for page in sorted_pages]
    return sorted_pages

# Function to retrieve indexed pages from Firestore
def get_indexed_pages1(query, session):
    # Check if the query result is cached
    if query in cache:
        if (datetime.now() - cache[query]['timestamp']).total_seconds() < CACHE_EXPIRATION:
          session_data = {
              "action": "search",
              "datetime": datetime.now(),
              "input": query,
              "output": cache[query]['data']
          }
          fbConn.post("/sessions/" + session, data=session_data)
          return cache[query]['data']

    # Fetch data from Firestore
    #indexed_pages = fbConn.get('/index/' + query, None)
    words = query.split()
    combined_counts = {}
    for word in words:
      results_for_word = fbConn.get('/index/' + word, None)
      if results_for_word:
        for hash_key, page_info in results_for_word.items():
            if hash_key in combined_counts:
                combined_counts[hash_key]['count'] += page_info['count']
            else:
                combined_counts[hash_key] = {'count': page_info['count']}
    link_hashes = []
    if combined_counts:
        link_hashes = sort_pages(combined_counts)

    # Update cache with query result
    cache[query] = {
        'data': link_hashes,
        'timestamp': datetime.now()
    }

    # Add the query and response to a session
    session_data = {
        "action": "search",
        "datetime": datetime.now(),
        "input": query,
        "output": link_hashes
    }
    fbConn.post("/sessions/" + session, data=session_data)

    return link_hashes

# Function to retrieve pages info from Firestore
def get_pages_info1(hashed_urls):
    links = []
    for hashed_url,count in hashed_urls:
        if hashed_url is not None:
            page_data = fbConn.get('/pages/' + hashed_url, None)
            if page_data:
                links.append((page_data,count))
    return links
```
