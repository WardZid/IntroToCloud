```python
import pandas as pd
file_path = '/content/drive/MyDrive/Intro to cloud/students.txt'

with open(file_path, 'r') as file:
    #content = file.read()
    df = pd.read_csv(file_path, delimiter=';')

# Display the DataFrame
df
#print(content)
#file.close()
```

```python
# @title ## Pick Student (run once and use the widgets)
import pandas as pd
import ipywidgets as widgets
from IPython.display import display


dropdown = widgets.Dropdown(
    options=df['FirstName'].tolist(),
    value=df['FirstName'].iloc[0],
    description='Select a student:'
)
course_text = widgets.Text(description='Courses:')
update_button = widgets.Button(description="Update and Save")

display(dropdown,course_text,update_button)


def on_dropdown_change(change):
    selected_name = change.new
    selected_row = df[df['FirstName'] == selected_name]
    course_text.value = selected_row['Courses'].iloc[0]


def on_text_change(change):
    selected_name = dropdown.value
    # record the change
    df.loc[df['FirstName'] == selected_name, 'Courses'] = course_text.value

def on_button_click(b):
    #save to my drive as a csv style file
    df.to_csv(file_path, sep=';', index=False)


dropdown.observe(on_dropdown_change, names='value')
course_text.observe(on_text_change, names='value')
update_button.on_click(on_button_click)

#to trigger the onChange event
dropdown.value = df['FirstName'].iloc[1]
dropdown.value = df['FirstName'].iloc[0]
```
